```python
df = pd.read_csv('dataset_ML_govvies_important1.csv')

X_train, Y_train, X_val, Y_val, X_test, Y_test = splitAll(df, non_X_cols = ['Date','Bond','Return_PD'], standardize= 'standard')
```

# Import Values

```python
Y_pred_OLS_all = pd.read_csv('Y_pred_OLS_all.csv')
Y_pred_OLS_4 = pd.read_csv('Y_pred_OLS_4.csv')
Y_pred_ENet = pd.read_csv('Y_pred_ENet.csv')
Y_pred_SVR = pd.read_csv('Y_pred_SVR.csv')
Y_pred_GLM = pd.read_csv('Y_pred_GLM.csv')
Y_pred_RF = pd.read_csv('Y_pred_RF.csv')
Y_pred_XGBoost = pd.read_csv('Y_pred_XGBoost.csv')
Y_pred_SVR = pd.read_csv('Y_pred_SVR.csv')
Y_pred_NN = pd.read_csv('Y_pred_NN.csv')
```

```python
df_pred = pd.DataFrame(columns= ('OLS all', 'OLS 4', 'Elastic Net', 'GLM', 'RF', 'XGBoost', 'SVR', 'NN'))

df_pred['OLS all'] = Y_pred_OLS_all
df_pred['OLS 4'] = Y_pred_OLS_4
df_pred['Elastic Net'] = Y_pred_ENet
df_pred['GLM'] = Y_pred_GLM
df_pred['RF'] = Y_pred_RF
df_pred['XGBoost'] = Y_pred_XGBoost
df_pred['SVR'] = Y_pred_SVR
df_pred['NN'] = Y_pred_NN
```

# Computing DM test stats

```python
dm_table = pd.DataFrame(index=df_pred.columns, columns=df_pred.columns)


for pair in itertools.combinations(df_pred.columns, 2):

    model1_pred = df_pred[pair[0]].to_numpy()
    model2_pred = df_pred[pair[1]].to_numpy()
    
    dm_stat = DM_adjusted(model1_pred, model2_pred, Y_test)
    dm_table.at[pair[0], pair[1]] = dm_stat

# Positive (Negative) values means the model on the column (row) outperformed
dm_table
```
