# Get data

```python
df = pd.read_csv('dataset_ML_govvies_important1.csv')
X_train, Y_train, X_val, Y_val, X_test, Y_test = splitAll(df, non_X_cols = ['Date','Bond','Return_PD'], standardize= 'minmax')
```

# XGB

```python
def evaluate_model_handwritten(Y_pred, Y_test):
    # R squared score

    num = (sum(pow((Y_test.ravel()-Y_pred),2)))
    denum = sum(pow(Y_test.ravel(),2))

    r2_test = 1-num/denum
    return r2_test
```

```python
params = {
'max_depth': [1],
'n_estimators': [100],
'learning_rate': [0.1],
'reg_alpha': [0.2],
}

r2_score = -100

for depth in params['max_depth']:
    for n_estimator in params['n_estimators']:
        for lr in params['learning_rate']:
            for reg_alpha in params['reg_alpha']:
                model = XGBRegressor(max_depth = int(depth), n_estimators = int(n_estimator), learning_rate = lr, reg_alpha = reg_alpha,)
                X_train_val = np.concatenate((X_train, X_val), axis=0)
                Y_train_val = np.concatenate((Y_train, Y_val), axis=0)
                model.fit(X_train_val, Y_train_val)
                Y_pred = model.predict(X_test)
                r2_test = evaluate_model_handwritten(Y_pred, Y_test)
                print(r2_test)
                if r2_test > r2_score:
                    r2_score = r2_test
```

```python
# XGBoost param grid
param_grid = {
'max_depth': [1],
'n_estimators': [100],
'learning_rate': [0.1],
'reg_alpha': [0.2],
}

Y_pred_ols_mf = fit_model(X_train, Y_train, X_val, Y_val, X_test, Y_test, 'XGB', param_grid, retain_month = 12, hypertuneOnce = False, static = True, pca_on = False)
```
