# Get data

```python
df = pd.read_csv('dataset_ML_govvies_important1.csv')
X_train, Y_train, X_val, Y_val, X_test, Y_test = splitAll(df, non_X_cols = ['Date','Bond','Return_PD'], standardize= 'minmax')
```

```python
# XGBoost param grid
param_grid = {
'max_depth': [1],
'n_estimators': [100],
'learning_rate': [0.1],
'reg_alpha': [0.2],
}

fit_model_static(X_train, Y_train, X_val, Y_val, X_test, Y_test, 'XGB', param_grid)
```

# Dynamic, annually retraining, no rehypertuning

```python
param_grid = {
'max_depth': [1],
'n_estimators': [100],
'learning_rate': [0.1],
'reg_alpha': [0.15],
}

Y_pred_xgb = fit_model(X_train, Y_train, X_val, Y_val, X_test, Y_test, 'XGB', param_grid, retain_month = 12, hypertuneOnce = True)
df1 = pd.DataFrame(Y_pred_xgb, columns = ['XGB_Y_pred'])
df1.to_csv('Y_pred_XGB.csv', index=False)
```
