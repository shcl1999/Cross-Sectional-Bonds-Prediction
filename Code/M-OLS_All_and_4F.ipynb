# Function

```python
def evaluate_model_handwritten(Y_pred, Y_test):
    # R squared score

    num = (sum(pow((Y_test-Y_pred),2)))
    denum = sum(pow(Y_test,2))

    r2_test = 1-num/denum
    return r2_test
```
